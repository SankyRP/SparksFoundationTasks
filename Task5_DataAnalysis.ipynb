# Grouping, Sorting unique values and plotting graphs 

```python
#Number of sales by each category
highest_sales_category = df['Category'].value_counts()
```

```python
# Graph for number of sales by each category
plt.figure(figsize =(20, 5))
plt.pie(highest_sales_category)
plt.xlabel('Category')
plt.ylabel('Number Of Sales')
plt.legend(['Office supplies', 'Furniture', 'Other'], bbox_to_anchor=(1, 1))
plt.title('Highest number of sales by each category - Graph 1', fontweight='bold', pad=30)
```

```python
#Number of sales by each State
highest_sales_states = df['State'].value_counts()
```

```python
#Graph for number of sales by each State
plt.figure(figsize =(20, 5))
plt.plot(highest_sales_states, color='red', linewidth=4)
plt.xlabel('State',  fontweight='bold')
plt.ylabel('Number Of Sales',  fontweight='bold')
plt.xticks(rotation=90)
plt.title('Number of sales by each State - Graph 2', fontweight='bold', pad=30, fontsize='larger')
```

```python
plt.figure(figsize =(15, 10))
y = df['Discount']
x = df['Profit']
plt.scatter(x, y, color='green')
plt.xlabel('Profits', fontweight='bold')
plt.ylabel('Discounts', fontweight='bold')
plt.title('Profits vs Discounts - Graph 3', fontweight='bold', pad=30, fontsize='larger')
```

```python
plt.figure(figsize =(20, 10))
y = df['Sales']
x = df['Sub-Category']
plt.bar(x, y)
plt.xlabel('Sub-Category', fontsize='large', fontweight='bold')
plt.ylabel('Sales', fontsize='large', fontweight='bold')
plt.title('Number of sales by each Sub-Category - Graph 4', fontsize='large', pad=30, fontweight='bold')
plt.show()
```

# Inference

1. Graph 1: By the pie chart we can infer that 'Office supplies' has the highest number of sales compared to the other categories
2. Graph 2: By the line chart we can infer the number of sales trend by each state with 'California' being the higest and 'Wyoming' being the least
3. Graph 3 - In the discount vs Profit graph, the highest number of profit is in the 0-2000 category irrespective of the discount provided.
4. Graph 4 - This graph tells us the trend of sale in each Sub-Category with 'Machines' being the highest and 'Fasteners' being the least
